```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import lit
spark = SparkSession.builder.appName('NSL-KDD').getOrCreate()
spark.sparkContext.setLogLevel("ERROR")
data = spark.read.csv('gs://bucket_nslkdd_hw4/KDDTest+.txt', header=False, inferSchema=True)
protocol_login_count = data.groupBy("_c1", "_c40").count()
print("Q1: Count of each protocol_type for logged in and non-logged in users")
protocol_login_count.show()
```
